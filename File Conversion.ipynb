```python
# Convert Json into CSV files
import json
import csv

def json_to_csv(json_filepath, csv_filepath):
    with open(json_filepath, 'r') as json_file, open(csv_filepath, 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        headers_written = False
        
        for line in json_file:
            try:
                data = json.loads(line)
            except json.JSONDecodeError:
                continue  # skips any lines that aren't valid JSON

            # Write headers once
            if not headers_written:
                headers = data.keys()
                csv_writer.writerow(headers)
                headers_written = True

            # Write data rows
            csv_writer.writerow(data.values())

# Replace 'path_to_json.json' and 'output.csv' with your actual file paths
json_to_csv('Business.json', 'Business.csv')
```
